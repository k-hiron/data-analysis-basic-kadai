```python
category_df = pd.read_csv('category.csv')
category_df
```

```python
df = pd.merge(df, category_df, on='商品番号', how='left')
```

```python
import matplotlib.pyplot as plt

# Now you can use the plot function
df['カテゴリー'].value_counts().plot(kind='bar')
```

```python
df.rename(columns={'カテゴリー': 'カテゴリー_y'}, inplace=True)

df = pd.merge(df, category_df[['商品番号', 'カテゴリー']], how='inner', on='商品番号')

df

plt.show()
```

```python
# 商品番号ごとの注文数のカウント
print(df.groupby('商品番号')['注文数'].value_counts())

# 商品番号ごとの注文数の基本統計量
print(df.groupby('商品番号')['注文数'].describe())
```
